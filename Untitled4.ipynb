```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Parameters
n_deliveries = 5000  # Expanded size
start_date = datetime(2025, 8, 10)
courier_types = ['Motorcycle', 'Van', 'Economic']
traffic_levels = ['Low', 'Medium', 'High']
cities = ['Tehran', 'Mashhad', 'Isfahan', 'Shiraz', 'Tabriz']  # New: Cities
city_weights = [0.4, 0.2, 0.15, 0.15, 0.1]  # Weighted for realism (Tehran = 40%)

# Generate data
data = {
    'delivery_id': range(1, n_deliveries + 1),
    'delivery_time': np.random.normal(loc=30, scale=10, size=n_deliveries).clip(10, 60),  # Normal dist, 10-60 min
    'distance': np.random.uniform(1, 20, n_deliveries),  # 1-20 km
    'cost': np.random.uniform(5, 50, n_deliveries),  # 5-50 units
    'courier_type': [random.choice(courier_types) for _ in range(n_deliveries)],
    'traffic_level': [random.choice(traffic_levels) for _ in range(n_deliveries)],
    'success': np.random.choice([0, 1], size=n_deliveries, p=[0.1, 0.9]),  # 90% success rate
    'city': np.random.choice(cities, size=n_deliveries, p=city_weights),  # Weighted city selection
    'timestamp': [start_date + timedelta(hours=random.randint(0, 24*7)) for _ in range(n_deliveries)]
}

# Create DataFrame
df = pd.DataFrame(data)

# Adjust delivery_time and cost based on traffic, courier, and city
for i, row in df.iterrows():
    # Traffic adjustments
    if row['traffic_level'] == 'High':
        df.at[i, 'delivery_time'] *= 1.5  # 50% slower
        df.at[i, 'cost'] *= 1.2  # 20% more expensive
    elif row['traffic_level'] == 'Medium':
        df.at[i, 'delivery_time'] *= 1.2
    # Courier adjustments
    if row['courier_type'] == 'Van':
        df.at[i, 'cost'] *= 1.5  # Vans cost more
    elif row['courier_type'] == 'Economic':
        df.at[i, 'cost'] *= 0.8  # Economic is cheaper
        df.at[i, 'delivery_time'] *= 1.3  # But slower
    # City adjustments (e.g., Tehran has higher costs due to urban density)
    if row['city'] == 'Tehran':
        df.at[i, 'cost'] *= 1.1  # 10% costlier in Tehran

# Ensure no negative values
df['delivery_time'] = df['delivery_time'].clip(10, 60)
df['cost'] = df['cost'].clip(5, 50)

# Simulate 5% missing data for delivery_time
missing_indices = random.sample(list(df.index), int(n_deliveries * 0.05))
df.loc[missing_indices, 'delivery_time'] = np.nan

# Save to CSV for reuse
df.to_csv('snappbox_deliveries_expanded.csv', index=False)

# Preview
print("First 5 Rows:")
print(df.head())
print("\nData Info:")
print(df.info())
print("\nCity Distribution:")
print(df['city'].value_counts(normalize=True))
```

```python
import pandas as pd
from pandasql import sqldf

# Load data
df = pd.read_csv('snappbox_deliveries_expanded.csv')

# Define SQL query function
pysqldf = lambda q: sqldf(q, globals())

# Query 1: Average delivery time and cost by city
query1 = """
SELECT city,
       AVG(delivery_time) AS avg_delivery_time,
       AVG(cost) AS avg_cost
FROM df
GROUP BY city;
"""
print("\nQuery 1: Average Delivery Time and Cost by City")
print(pysqldf(query1))

# Query 2: Success rate by courier type
query2 = """
SELECT courier_type,
       AVG(success) AS success_rate
FROM df
GROUP BY courier_type;
"""
print("\nQuery 2: Success Rate by Courier Type")
print(pysqldf(query2))

# Query 3: Deliveries with high traffic and time > 50 min (potential delays)
query3 = """
SELECT city, courier_type, COUNT(*) AS high_delay_count
FROM df
WHERE traffic_level = 'High' AND delivery_time > 50
GROUP BY city, courier_type;
"""
print("\nQuery 3: High Traffic Deliveries with Time > 50 min")
print(pysqldf(query3))

# Query 4: Average delivery time by city and traffic level
query4 = """
SELECT city, traffic_level,
       AVG(delivery_time) AS avg_delivery_time
FROM df
GROUP BY city, traffic_level
ORDER BY city, avg_delivery_time DESC;
"""
print("\nQuery 4: Average Delivery Time by City and Traffic Level")
print(pysqldf(query4))

# Query 5: Missing delivery_time analysis
query5 = """
SELECT city,
       COUNT(*) AS missing_count
FROM df
WHERE delivery_time IS NULL
GROUP BY city;
"""
print("\nQuery 5: Missing Delivery Time Count by City")
print(pysqldf(query5))
```

```python
import pandas as pd
import json
from datetime import datetime

# Simulate Redis with a Python dictionary
redis_cache = {}

# ETL Pipeline
def extract_data(file_path):
    """Extract data from CSV."""
    print("Extracting data...")
    df = pd.read_csv(file_path)
    print(f"Extracted {len(df)} records.")
    return df

def transform_data(df):
    """Transform data: clean, aggregate, flag delays."""
    print("Transforming data...")

    # Clean missing values
    df['delivery_time'] = df['delivery_time'].fillna(df['delivery_time'].median())

    # Flag high-delay deliveries (>50 min)
    df['high_delay'] = df['delivery_time'] > 50

    # Aggregate metrics by city and courier_type
    summary = df.groupby(['city', 'courier_type']).agg({
        'delivery_time': 'mean',
        'cost': 'mean',
        'success': 'mean',
        'high_delay': 'sum'
    }).reset_index()
    summary = summary.rename(columns={
        'delivery_time': 'avg_delivery_time',
        'cost': 'avg_cost',
        'success': 'success_rate',
        'high_delay': 'high_delay_count'
    })

    print("Transformation complete.")
    return df, summary

def load_data(df, summary, output_file, cache_key):
    """Load data to CSV and cache."""
    print("Loading data...")

    # Save summary to CSV
    summary.to_csv(output_file, index=False)
    print(f"Saved summary to {output_file}")

    # Cache metrics in 'Redis' (dictionary)
    redis_cache[cache_key] = summary.to_dict(orient='records')
    print(f"Cached metrics under key: {cache_key}")

    # Simulate real-time access
    print("\nSample Cached Data (First Record):")
    print(json.dumps(redis_cache[cache_key][0], indent=2))

    return df, summary

def run_pipeline(file_path, output_file, cache_key):
    """Run the full ETL pipeline."""
    print("Starting ETL pipeline...")
    df = extract_data(file_path)
    df, summary = transform_data(df)
    df, summary = load_data(df, summary, output_file, cache_key)
    print("Pipeline complete.")
    return df, summary

# Run pipeline
if __name__ == "__main__":
    file_path = 'snappbox_deliveries_expanded.csv'
    output_file = 'snappbox_summary.csv'
    cache_key = f"delivery_metrics_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    df, summary = run_pipeline(file_path, output_file, cache_key)

    # Preview results
    print("\nSummary Output (First 5 Rows):")
    print(summary.head())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# Load data
df = pd.read_csv('snappbox_deliveries_expanded.csv')
df['delivery_time'] = df['delivery_time'].fillna(df['delivery_time'].median())

# Split into Control (A) and Test (B)
np.random.seed(42)
df['group'] = np.random.choice(['Control', 'Test'], size=len(df), p=[0.5, 0.5])
control = df[df['group'] == 'Control']['delivery_time']
test = df[df['group'] == 'Test']['delivery_time'].copy()

# Simulate optimized algorithm (10% reduction in Test group)
test *= 0.9  # Reduce delivery time by 10%

# T-test
t_stat, p_value = ttest_ind(control, test, equal_var=False)
print("A/B Test Results:")
print(f"Control Mean Delivery Time: {control.mean():.2f} minutes")
print(f"Test Mean Delivery Time: {test.mean():.2f} minutes")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Significant difference! Optimized algorithm reduces delivery time.")
else:
    print("No significant difference.")

# City-level analysis
city_results = []
for city in df['city'].unique():
    control_city = df[(df['group'] == 'Control') & (df['city'] == city)]['delivery_time']
    test_city = df[(df['group'] == 'Test') & (df['city'] == city)]['delivery_time'] * 0.9
    if len(control_city) > 10 and len(test_city) > 10:  # Ensure enough samples
        t_stat_city, p_value_city = ttest_ind(control_city, test_city, equal_var=False)
        city_results.append({
            'City': city,
            'Control Mean': control_city.mean(),
            'Test Mean': test_city.mean(),
            'P-value': p_value_city
        })
city_results = pd.DataFrame(city_results)
print("\nCity-Level A/B Test Results:")
print(city_results)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Load data (from your A/B test setup)
df = pd.read_csv('snappbox_deliveries_expanded.csv')
df['delivery_time'] = df['delivery_time'].fillna(df['delivery_time'].median())
np.random.seed(42)
df['group'] = np.random.choice(['Control', 'Test'], size=len(df), p=[0.5, 0.5])
control = df[df['group'] == 'Control']['delivery_time']
test = df[df['group'] == 'Test']['delivery_time'] * 0.9  # 10% reduction

# Calculate Cohen's d
def cohens_d(control, test):
    mean_diff = control.mean() - test.mean()
    pooled_std = np.sqrt((control.std()**2 + test.std()**2) / 2)
    return mean_diff / pooled_std

overall_d = cohens_d(control, test)
print("Overall Cohen's d:", f"{overall_d:.2f}")

# City-level Cohen's d
city_results = []
for city in df['city'].unique():
    control_city = df[(df['group'] == 'Control') & (df['city'] == city)]['delivery_time']
    test_city = df[(df['group'] == 'Test') & (df['city'] == city)]['delivery_time'] * 0.9
    if len(control_city) > 10 and len(test_city) > 10:
        t_stat_city, p_value_city = ttest_ind(control_city, test_city, equal_var=False)
        d = cohens_d(control_city, test_city)
        city_results.append({
            'City': city,
            'Control Mean': control_city.mean(),
            'Test Mean': test_city.mean(),
            'P-value': p_value_city,
            'Cohen\'s d': d
        })
city_results = pd.DataFrame(city_results)
print("\nCity-Level A/B Test Results with Effect Size:")
print(city_results)
```
